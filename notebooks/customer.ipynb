```python
# Meraj Nushin
# Customer Churn Analysis

import pandas as pd

customerDf = pd.read_csv('/Users/merajnushin/Downloads/Customer-Churn-Analysis/csv/customer-churn.csv')

# Display the first few rows of the DataFrame
customerDf.head()
```

```python
# Display information about the DataFrame of the customer data
customerDf.info()
```

```python
# Display the last few rows of the DataFrame
customerDf.tail()
```

```python
# Display the summary statistics of the DataFrame
customerDf.describe()
```

```python
# Convert 'TotalCharges' to numeric, forcing errors to NaN, then drop rows with NaN values

customerDf['TotalCharges'] = pd.to_numeric(customerDf['TotalCharges'], errors='coerce')
customerDf.dropna(inplace=True)

customerDf.info()
```

```python
# Calculate and display the churn rate
churn_rate = customerDf['Churn'].value_counts(normalize=True)
print(churn_rate)

# Explanatory Data Analysis (EDA)
# What is the churn rate among customers?
# Which customers groups churn more frequently?
# Does contract type affect churn rate?
```

```python
# Visualizing Monthly Charges by Churn Status

import seaborn as sns 
import matplotlib.pyplot as plt
sns.boxplot(x='Churn', y='MonthlyCharges', data=customerDf)
plt.show()
```

```python
# Month-to-month customers churn more
sns.countplot(x='Contract', hue='Churn', data=customerDf)
plt.show()
```

```python
# Higher monthly charges lead to higher churn
sns.histplot(x='MonthlyCharges', hue='Churn', data=customerDf, multiple='stack', bins=30)
plt.show()
```

```python
# longer tenure customers churn less
sns.histplot(x='tenure', hue='Churn', data=customerDf, multiple='stack', bins=30)
plt.show()
```

```python
# Predictive Modeling Using Logistic Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = customerDf[['MonthlyCharges', 'tenure']]
y = customerDf['Churn'].map({'Yes': 1, 'No': 0})

X_train, X_test, y_train, y_test = train_test_split(X, y)

model = LogisticRegression()
model.fit(X_train, y_train)
accuracy = model.score(X_test, y_test)
print(f'Model Accuracy: {accuracy:.2f}')
```
